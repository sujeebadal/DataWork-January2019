```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("datasets/DC_Properties.csv", encoding = "ISO-8859-1") 
df.head()
```

```python
#make a scatterplot
sns.lmplot(x='ROOMS', y='PRICE', data=df) 
```

```python
#make a scatterplot
sns.lmplot(x='YR_RMDL', y='PRICE', data=df) 
```

```python
df2= df [['ZIPCODE','AYB','EYB','YR_RMDL','SALEDATE','BEDRM','ROOMS','LIVING_GBA','PRICE']]
df2
```

```python
df3= df2.loc[(df2['PRICE'] >= 3000000) & (df2['PRICE'] < 10000000)]
df3
```

```python
#make a scatterplot
sns.lmplot(x='ZIPCODE', y='PRICE', data=df3)
```

```python
#make a scatterplot
sns.lmplot(x='YR_RMDL', y='PRICE', data=df3) 
```

```python
#add more details to scatterplot
sns.lmplot(x='ZIPCODE', y='PRICE', data=df, 
           fit_reg=False, #remove regression line
           hue='ZIPCODE')   #color by ZIPCODE
```

```python
#boxplot showing stats for each characteristic
sns.boxplot(data=df3)
```

```python
#DROP VARIABLES
stats_df3 = df3.drop(['YR_RMDL','SALEDATE','BEDRM','LIVING_GBA'], axis=1)
stats_df3.head(577)
```

```python
#boxplot for stats
sns.boxplot(data=stats_df3)
```

```python
#change size of plot
plt.subplots(figsize=(13,7))

#violin plot shows the distribution of PRICE for each ROOM
sns.violinplot(x='ROOMS', y='PRICE', data=df3)
```

```python
#create a histogram of the distribution of PRICE
sns.distplot(df3['PRICE']) 
```

```python
#create a bar plot of Type 1 frequency
sns.countplot(x='ZIPCODE',
              data=df3)

plt.xticks(rotation=-45)
```
